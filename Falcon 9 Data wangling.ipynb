# Orbital Types and Definitions

- **LEO (Low Earth Orbit)**  
  An Earth-centred orbit with an altitude of 2,000 km (1,200 mi) or less (approximately one-third of the radius of Earth), or with at least 11.25 periods per day (an orbital period of 128 minutes or less) and an eccentricity less than 0.25.  
  Most manmade objects in outer space are in LEO.

- **VLEO (Very Low Earth Orbit)**  
  Orbits with a mean altitude below 450 km. Operating in these orbits provides benefits to Earth observation spacecraft by being closer to the observation targets. 

- **GTO (Geosynchronous Transfer Orbit)**  
  A high Earth orbit that allows satellites to match Earth's rotation, located at 22,236 miles (35,786 km) above Earth's equator. Satellites here appear to stay fixed over a single longitude, useful for weather, communications, and surveillance. 

- **SSO (Sun-Synchronous Orbit)**  
  Also called a heliosynchronous orbit, it is a nearly polar orbit where the satellite passes over any given point of the planet’s surface at the same local mean solar time. 

- **ES-L1 (Earth-Sun Lagrange Point 1)**  
  A point where gravitational forces between the Sun and Earth cancel out, allowing a small object to remain in equilibrium relative to the two bodies. L1 is such a point between the Sun and Earth. 

- **HEO (Highly Elliptical Orbit)**  
  An elliptic orbit with high eccentricity, usually referring to one around Earth. 

- **ISS (International Space Station Orbit)**  
  A modular, habitable artificial satellite in low Earth orbit. A multinational collaborative project involving NASA, Roscosmos, JAXA, ESA, and CSA. 

- **MEO (Medium Earth Orbit)**  
  Geocentric orbits ranging from 2,000 km up to just below geosynchronous orbit (35,786 km). Also called intermediate circular orbits, most commonly at ~20,200 km altitude with a 12-hour orbital period. 

- **HEO (Geocentric High Earth Orbit)**  
  Geocentric orbits above the altitude of geosynchronous orbit (above 35,786 km). 

- **GEO (Geosynchronous Earth Orbit)**  
  A circular geosynchronous orbit 35,786 km above Earth’s equator, following Earth's rotation. 

- **PO (Polar Orbit)**  
  Orbits in which a satellite passes above or nearly above both poles of the body it orbits (usually Earth).
